```python
import pandas #type: ignore
from pandas import DataFrame
from matplotlib import pyplot as plot
from sklearn.linear_model import LinearRegression
```

```python
cost_data = pandas.read_csv("cost_revenue_clean_1.csv")
cost_data.describe()
```

```python
X = DataFrame(cost_data, columns=["Production_Budget"])
Y = DataFrame(cost_data, columns=["Worldwide_Gross"])
```

```python
X, Y
```

```python
plot.figure(figsize=(10,6))
plot.scatter(X,Y,alpha=0.3)
plot.ylabel("Production Budget")
plot.xlabel("Worldwide Gross")
plot.ylim(0, 3000000000)
plot.xlim(0, 450000000)
plot.show()
```

Linear regration 

```python
regression = LinearRegression()
regression.fit(X, Y)
```

```python
regression.coef_
```

```python
regression.intercept_
```

```python
plot.figure(figsize=(10,6))
plot.scatter(X,Y,alpha=0.3)
plot.plot(X, regression.predict(X), color="red", linewidth=3)

plot.ylabel("Production Budget")
plot.xlabel("Worldwide Gross")
plot.ylim(0, 3000000000)
plot.xlim(0, 450000000)
plot.show()
```

```python
regression.score(X, Y)
```
